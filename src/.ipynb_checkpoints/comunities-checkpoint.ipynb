```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats
from sklearn.model_selection import train_test_split
from sklearn.cluster import KMeans
from sklearn.preprocessing import MinMaxScaler
import csv

sns.set(color_codes=True)

pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)

from IPython.core.display import display, HTML
display(HTML("<style>.container { width:100% !important; }</style>"))

# read data
data = pd.read_csv('data/Modelar_UH2020.txt', sep="|")
```

##### VARIABLES

La información de longitud-latitud ha sido escalada y desplazada aleatoriamente

- X: Longitud (se matiene la relación de posición con el resto de puntos)
- Y: Latitud (se matiene la relación de posición con el resto de puntos)

```python
from sklearn.preprocessing import LabelEncoder

# creating instance of labelencoder
labelencoder = LabelEncoder()

# Assigning numerical values and storing in another column
data['label_encoder'] = labelencoder.fit_transform(data['CLASE'])
data[['CLASE','label_encoder']].drop_duplicates()
```

```python
data.plot(kind="scatter", x="X", y="Y", alpha=0.4,figsize=(20,15),c="label_encoder",colorbar=True,cmap=plt.get_cmap("jet"))
plt.show()
```

##### Clustering X-Y

```python
# Split train-test
train, test = train_test_split(data, test_size=0.2, random_state=123)

X_train = train.loc[:, train.columns != 'CLASE']
X_test = test.loc[:, test.columns != 'CLASE']
y_train = train['CLASE']
y_test = test['CLASE']
```

```python
# Normalize XY train/test set
scaler = MinMaxScaler()
scaler.fit(X_train.loc[:,['X','Y']])

X_train = X_train.copy()
X_test = X_train.copy()

X_train.loc[:,['X','Y']] = scaler.transform(X_train.loc[:,['X','Y']])
X_test.loc[:,['X','Y']] = scaler.transform(X_test.loc[:,['X','Y']])
```

```python
# Elbow curve
K_clusters = range(1,20)
kmeans = [KMeans(n_clusters=i) for i in K_clusters]
score = [kmeans[i].fit(X_train.loc[:,['X','Y']]).score(X_train.loc[:,['X','Y']]) for i in range(len(kmeans))]
# Visualize
plt.plot(K_clusters, score)
plt.xlabel('Number of Clusters')
plt.ylabel('Score')
plt.title('Elbow Curve')
plt.show()
```

```python
# Clustering using K-Means and Assigning Clusters to our Data
kmeans = KMeans(n_clusters = 15, init ='k-means++')
kmeans.fit(X_train.loc[:,['X','Y']]) # Compute k-means clustering.

X_train['comunity'] = kmeans.fit_predict(X_train.loc[:,['X','Y']])
X_test['comunity'] = kmeans.fit_predict(X_test.loc[:,['X','Y']])

centers = pd.DataFrame(kmeans.cluster_centers_).reset_index().rename(columns={'index':'comunity',0:'X_comunity',1:'Y_comunity'}) # Coordinates of cluster centers.
```

```python
# Merge
X_train = pd.merge(X_train,centers,'left','comunity')
X_test = pd.merge(X_test,centers,'left','comunity')

# Distancia euclidea a su comunidad
X_train['ecludina_dist'] = np.sqrt((X_train.X_comunity-X_train.X)**2+(X_train.Y_comunity-X_train.Y)**2)
X_test['ecludina_dist'] = np.sqrt((X_test.X_comunity-X_test.X)**2+(X_test.Y_comunity-X_test.Y)**2)
```

```python
plt.figure(figsize=(20,20))
plt.scatter(x = X_train.loc[:,'X'], y = X_train.loc[:,'Y'], c=X_train.loc[:,'comunity'], s=50, cmap='viridis')
plt.scatter(centers.loc[:,'X_comunity'], centers.loc[:,'Y_comunity'], c='black', s=200, alpha=0.5)
plt.savefig('clust.png')

plt.figure(figsize=(20,20))
plt.scatter(x = X_train.loc[:,'X'], y = X_train.loc[:,'Y'], c=X_train.loc[:,'label_encoder'], s=50, cmap='viridis')
plt.scatter(centers.loc[:,'X_comunity'], centers.loc[:,'Y_comunity'], c='black', s=200, alpha=0.5)
plt.savefig('class.png')
```
